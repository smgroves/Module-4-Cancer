## Disease Background:

* Cancer Hallmark focus: Immune response evasion and metabolic reprogramming
* Overview of hallmark: 
    * Immune Response Evasion: Cancer cells are able to evade immune cells, such as Killer T cells, as they traverse through the human body and continue to develop uncontrollably. Due to this evasion, it makes it harder for the immune system to attack the cancer cells, causing the cancer to prevail for longer with minimal damage. This is a programming of the cancer cell as per its sense of dividing as much as possible.
    * Metabolic Reprogramming: Cancer cells are able to create systems to allow themselves to survive. A process they commonly take is glycolysis, which allows them to survive in oxygen-poor areas, which otherwise would've caused them to die if they didn't have the resources. This is a reprogramming of the cancer cell's abilities as to allow them to live longer, so that they can develop uncontrollably.
* Genes associated with hallmark: Immune evasion hallmark includes key oncogenes such as MYC, RAS, and STAT3, and tumor suppressors TP53 and PTEN, which are important for immune suppresion and to avoid detection by key immune cells. For metabolic reprogramming, oncogenes involved are MYC, KRAS, and PIK3CA, and tumor suppressors such as p53, LKB1/AMPK that improve nutrient usage.

### Background on Glioblastoma
Glioblastoma is an aggressive type of brain cancer that begins with astrocytes (support nerve cells). Specifically, glial cells - cells in the central and peripheral nervous system - that endure a major mutation/change in the DNA, which then causes them to fail within their own division processes, leading to glioblastoma. Glioblastoma is the most common malignant form of glioma, representing a wide array of glial cell-formed brain cancer.

It's the most common malignant brain tumor in adults and makes up about half of all malignant brain tumors in adults.

Within this group, glioblastoma is classified as a grade 4 glioma (Mayo Clinic, 2024a), the most serious type of glioma. Grade 4 means the tumor grows and spreads quickly into nearby brain tissue, which makes it hard to remove completely.

Symptoms:
Symptoms vary based on where the tumor forms in the brain and may involve seizures, brain swelling, headaches, personality changes, and changes in speech, vision, strength, sensation or balance.

There is no cure for glioblastoma. Treatments such as surgery, radiation and chemotherapy may slow tumor growth, ease symptoms and help people live longer. These are treatment methods that hope to put the cancer in remission by removing the mass, killing off cells that divide fast, and creating environments to where it is harder for such fast-dividing cells to live.

The cause of most glioblastomas is not known. A lot can be pointed to major mutations or transcription factor changes within glial cells that cause them to mishap when dividing as a means to support the brain's nutrition. Poor lifestyle choices compounded over years creates more effective breeding grounds for cancer as mistakes for cells are more prone, the body's immune response drops with time, and cells have accumulated damage from years of previous mistakes as well. 

Risk Factors: 
* Radiation exposure 
* Old age
* Being male
* Inherited syndromes like Lynch Syndrome and Li-Fraumeni syndrome

There are no known mechanisms as to prevent glioblastoma. A lot of general mechanisms point to public health awareness and lifestyle changes as being a deep factor, since that reduces the abilities for cells to drift, make changes, and have unnecessary mutations that would've taken place if not for proper cell management. However, people can still develop severe versions even when organized healthily, because biology's laws aren't limited.

### 1. Import The Necessary Libraries & Datasets

Import the necessary libraries required for the machine learning model and visualization, and read the .csv files using pandas to understand them better!

```python
# Import all necessary files.
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Read the two dataset files, the metadata and the subsample, as .csv using pandas for better readability and usability.
data = pd.read_csv(
    r"C:\Users\dance\OneDrive - University of Virginia\Computational BME\Module-4-Cancer\data\TRAINING_SET_GSE62944_subsample_log2TPM.csv", index_col=0, header=0)  # can also use larger dataset with more genes
metadata_df = pd.read_csv(
    r"C:\Users\dance\OneDrive - University of Virginia\Computational BME\Module-4-Cancer\data\TRAINING_SET_GSE62944_metadata.csv", index_col=0, header=0)
```

### 2. List The Data And Organize Variables

Our chosen cancer type is Glioblastoma (GBM), and so, we retrieved data regarding it, and identified the genes related to immune evasion and listed them as per the data from the Menyhart documentation. This is to set up so that all the data is loaded before we run the PCA and other analyses to start visualization.

```python
# Organize variables and list them with data points about the cancer we chose.
cancer_type = 'GBM' 
cancer_samples = metadata_df[metadata_df['cancer_type'] == cancer_type].index
GBM_data = data[cancer_samples]

# List all specific genes for immune evasion within the list so that they can be graphed
desired_gene_list = ['STAT1', 'CXCR4', 'PTPN6', 'RHOA', 'STAT3', 'LCK', 'CD86', 'MAPK14', 'HCK', 'PTK2', 'HIF1A', 'PRF1']
gene_list = [gene for gene in desired_gene_list if gene in GBM_data.index]
```

### 3. Plot Pair Grid Graphs

We now plot pair grid graphs using the seaborn module to see relationships with specific variables within the datasets. This gives us a full view for all the genes as well as how they relate to the patient samples all in their own way.

Essentially, we labelled the data for GBM and the 12 genes chosen and doing Pair Grid to show correlation of the genes to the patient samples.

```python
# Set up all the variables for the pair grid graphs so that they can be plotted.
X = GBM_data.loc[gene_list]
y = gene_list
df_pair = pd.DataFrame(X.iloc[:, :4]) 
df_pair['Gene'] = y
# The pair grid will draw 16 graphs with the individual variables.
g = sns.PairGrid(df_pair, hue="Gene", palette="tab20")
g.map(sns.scatterplot)
plt.show()
```

### 4. Conduct PCA (Principal Component Analysis)

We now conduct the PCA onto the data, which uses dimensionality reduction to simplify the data, but still maintain the level of variance that existed within the original data. This helps us simplify the relationships with the genes whilst still seeing levels of accurate variance between all types.

```python
# Use standardscaler to scale the data points to the unit variance.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Use PCA to simplify the data and maintain the variance, and write to the respective variables.
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

Based on the collected values from the PCA functions, plot the PCA and show it.

```python
# Plot a scatterplot of the data points that underwent the PCA function.
sns.scatterplot(x=X_pca[:, 0], 
                y=X_pca[:, 1], 
                hue=y, #hue = y used to color the points based on gene labels
                palette="tab20", 
                s=200) # Adjusted this number to make the size of dots bigger

# Added relevant titles to the graphs for PCA.
plt.title("PCA for GBM using 12 Genes for Immune Response Invasion")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title="Genes")
plt.show()
```

### 5. Conduct Clustering

We now conduct 3 clustering mechanisms: K-Means, Hierarchical Clustering, and DBSCAN Clustering. All clustering mechanisms are key in helping us both visualize and understand the data better. 

Below is K-Means clustering, which assigns data points to specific groups based on the nearest centroids that exist to the data point.

```python
# Use a K-Means clustering system with 3 clusters for the scatterplot below.
model = KMeans(n_clusters=3, random_state=0)
model.fit(X_scaled)
y_pred = model.predict(X_scaled)

# Plot the scatter plot with the K-Means data.
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_pred, cmap="Set2", s=150)
plt.title("KMeans Clustering of Genes")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

Below is Hierarchical Clustering, which uses a tree-like structure to group similar data points. It's also known as a dendrogram.

The in-class notebook was used as a guide to help create the Hierarchical Clustering.

```python
# Plot the dendrogram of the similar genes
Z = linkage(X_scaled, method='ward')
plt.figure(figsize=(10, 5))
dendrogram(Z, labels=gene_list) # Label the tree with Gene Names
plt.title("Hierarchical Clustering of 12 Genes")
plt.show()
```

Below is the DBSCAN Clustering, which groups data points based on their density, doing a similar thing like K-Means to where it groups the data points.

The in-class notebook was used as a guide to help create the DBSCAN Clustering.

```python
dbscan = DBSCAN(eps=5.0, min_samples=2) # EPS adjusted for small sample size (12 points)
y_dbscan = dbscan.fit_predict(X_scaled)

# Plot the scatterplot with the DBSCAN data.
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_dbscan, cmap="Set2", s=150)
plt.title("DBSCAN Clustering of 12 Genes")
plt.show()
```
